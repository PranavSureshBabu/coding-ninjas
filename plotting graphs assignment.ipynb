```python
import pandas as pd
import matplotlib.pyplot as plt
```

# line plots

```python
df = pd.read_csv('amazon_jobs_dataset.csv')
df['Posting_date'].sample(15)
df['year'] = df['Posting_date'].str.split(',').str[1].astype('int')
x = df.groupby(['year']).count()['Title']
plt.plot(x.index, x.values)
plt.show()
for i in range (x.shape[0]):
    print(x.index[i], x.values[i])
```

# Bar graph

```python
df = pd.read_csv('amazon_jobs_dataset.csv')
df['month'] = df['Posting_date'].str.split(',').str[0].str.split(' ').str[0]
x = df.groupby(['month']).count()['Title']
plt.bar(x.index, x.values)
plt.xticks(rotation = 90)
plt.show()
for i in range (x.shape[0]):
    print(x.index[i], x.values[i])
```

# scatter plots

```python
df = pd.read_csv('amazon_jobs_dataset.csv')
java_df = df[df['BASIC QUALIFICATIONS'].str.contains('[Jj]ava') == True]
java_df['year'] = java_df['Posting_date'].str.split(',').str[1].astype('int')
x = java_df.groupby('year').count()['Title']
plt.scatter(x.index, x.values)

plt.show()

for i in range(x.shape[0]):
    print(x.index[i], x.values[i])
```

# pie chart

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('amazon_jobs_dataset.csv')
df['cities'] = df['location'].str.split(',').str[2]
df['country'] = df['location'].str.split(',').str[0] 
ind_df = df[df['country'] == 'IN']
x = ind_df.groupby(['cities']).count()['Title']
y = x.sort_values(ascending = False)
plt.pie(y.values, labels = y.index, autopct = '%.2f%%')
plt.show()
pct_list = [round(i/sum(y.values)* 100, 2) for i in y.values]
for i in range(y.shape[0]):
    print(y.index[i], format(pct_list[i],'2f'))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('amazon_jobs_dataset.csv')
location = data['location'].str.split(',', expand = True)
arr = location[0] == 'IN'
india_city = location[arr][2]
freq = india_city.value_counts()
plt.axis('equal')
x = freq.index
plt.pie(freq, labels = x, autopct = '%.2f%%')
plt.show()
perc = np.true_divide(freq, freq.sum(axis = 0))*100
for i in range(len(x)):
    print(x[i], format(perc[i], '2f'))
```
